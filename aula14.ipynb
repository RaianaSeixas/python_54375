### Exercícios

1 -Utilizando a base "realtor-data.csv" gerar uma base com os seguintes requisitos:

- Eliminar as linhas com missing values em variáveis categóricas.
- Substituir missing values das colunas bed e bath por zero.
- Substituir missing values das colunas house_size pela mediana

```python
df.info()
```

```python
df_tratado = df.dropna(subset = ['status', 'city', 'state', 'prev_sold_date']).reset_index(drop=True)
df_tratado = df_tratado.fillna({'bed':0,'bath':0, 'house_size': df_tratado['house_size'].median()})
df_tratado.head()
```

2. Da base anterior, criar a seguinte tabela pivot:

```python
df_tratado['prev_sold_date'] = pd.to_datetime(df_tratado['prev_sold_date'])
df_tratado['prev_sold_year'] = df_tratado['prev_sold_date'].dt.year
df_tratado
```

```python
pd.pivot_table(df_tratado,index = 'state', columns='prev_sold_year', values='price', aggfunc='sum')
```
