```python
dff=df.sort_values(['status', 'city', 'state', 'price','house_size' ],ascending=False)
dff
```

```python
dff = df.drop_duplicates(subset = ['status', 'city', 'state', 'price'])
dff
```

```python
dff = dff[df.status == 'for_sale']
dff
```

```python
mass = dff[dff['state']=='Massachusetts']
mass = mass.sample(n=250)
```

```python
# "Rhode Island", "New York"
combinado = dff[ dff['state'].isin(["Rhode Island", "New York"])].sample(n=250)
combinado
```

```python
df_sample = pd.concat([mass, combinado])
df_sample
```
