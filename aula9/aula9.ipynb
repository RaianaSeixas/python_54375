```python
# Conectar ao banco de dados
conn = sqlite3.connect('teste.db')
```

```python
# Salvar um dataframe no banco de dados
fastfood.to_sql('fastfood', conn, if_exists='replace', index = False)
```

```python
# Executar consulta no BD

query = "select item, calories from fastfood where calories > 500 and calories < 800 order by calories"

df = pd.read_sql(query, conn)
df
```

```python
def tabelas_bd():
    """
    retorna um dataframe com as tabelas do banco de dados.
    """
    conn = sqlite3.connect('coderhouse.db')
    query = "SELECT name FROM sqlite_master WHERE type='table'"
    schema = pd.read_sql_query(query, conn)
    conn.close()
    return schema
```

```python
tabelas_bd()
```

```python
def salva_bd(df, nome_tabela):
    """
    salva dataframe df na tabela chamada nome_tabela no banco de dados.
    """

    conn = sqlite3.connect('coderhouse.db')
    df.to_sql(nome_tabela, conn, if_exists='replace')
    conn.close()
    return None
```

```python
salva_bd(fastfood, 'fast')
```

```python
tabelas_bd()
```

```python
def carrega_bd(nome_tabela):
    """
    Carrega a tabela nome_tabela do banco de dados em um dataframe.
    """
    conn = sqlite3.connect('coderhouse.db')
    query = f"SELECT * from {nome_tabela}"
    df = pd.read_sql(query, conn)
    conn.close()
    return df
```

```python
carrega_bd('produtos')
```

```python
carrega_bd('fast')
```
