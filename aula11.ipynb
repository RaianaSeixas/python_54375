```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Tipos de Graficos
### Grafico dispersão

```python
x = [1, 2, 3, 4, 5]
y = [10, 8, 6, 4, 2]
```

### Gráfico de linhas

```python
plt.plot(x,y)
```

### Gráfico de barras

```python
xb = ['A', 'B', 'C', 'D', 'E']
yb = [3, 7, 1, 8, 4]
plt.bar(xb, yb)
plt.show()
```

### Gráfico de pizza

```python
plt.pie(yb, labels = xb)
plt.show()
```

```python
labels = ['Maçãs', 'Bananas', 'Laranjas', 'Morangos']
quantidades = [20, 25, 10, 30]
plt.pie(quantidades, labels=labels)
plt.show()
```

### Histograma

```python
dados = [1, 3, 3, 4, 4, 4, 5, 5, 6, 7, 8, 9]

plt.hist(dados, bins =4)
plt.title('Meu título')
plt.xlabel('eixo x')
plt.ylabel ('eixo y')
```

```python
sns.displot(dados, kde=True)
```

### Gráfico de área

```python
xa = [1, 2, 3, 4, 5]
ya = [2, 4, 6, 8, 10]
plt.fill_between (xa, ya)
```

## Configurações

```python
x = np.array([i for i in range(2,20,3)])
y = x **2 +1


# Configurações do gráfico

plt.plot(x, y, marker='o', color='red', markersize = 10, markerfacecolor = 'black', linewidth = 3, label = 'dados')
plt.title('Meu título')
plt.xlabel('eixo x')
plt.ylabel ('eixo y')
plt.grid(False)
#plt.xlim(0,25)

x2 =[10, 10, 10, 10, 10, 10]

plt.plot(x2,y, label = 'meta', color = 'black')
plt.legend()
plt.show()
```

## Subplot

```python
# Dados para os gráficos de linha
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)

# Criação de subplots de linhas

fig, axs = plt.subplots(3,1)
axs[0].plot(x, y1)
axs[0].set_title('Seno')
axs[1].plot(x, y2)
axs[2].plot(x, y3)
```

## Seaborn


```python
# Plotar scatter plot

data = {'peso': [80, 65, 70, 60, 75, 66, 61, 73, 77],
        'altura': [180, 170, 175, 167, 172, 169, 166, 178, 169]}
df = pd.DataFrame(data)

sns.scatterplot(x='peso', y='altura', data = df)
```

```python
df
```

```python
# Plotar boxplot

data = {'tipo': ['Venda', 'Venda', 'Compra', 'Compra', 'Venda', 'Compra'],
        'valor': [1000, 1500, 500, 800, 1200, 600]}

df = pd.DataFrame(data)
```

```python
df
```

```python
sns.set_theme(style='whitegrid') # Escolher entre temas pré-defnidos
sns.boxplot(x='tipo', y='valor', data = df, hue = 'tipo', palette = "Set1")
plt.show()
```
